```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('synthetic_loan_data.csv')
df.head()
```

## Plotting the data from the dataset

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MaritalStatus'], color='rebeccapurple')
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, color='rebeccapurple', kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
existing_loans_count = df['ExistingLoansCount'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=existing_loans_count.index, y=existing_loans_count.values, palette='magma')
plt.title('Distribution of Existing Loans Count')
plt.xlabel('Existing Loans Count')
plt.ylabel('Frequency')
plt.show()
```

```python
late_payments_last_year = df['LatePaymentsLastYear'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=late_payments_last_year.index, y=late_payments_last_year.values, palette='magma')
plt.title('Distribution of Late Payments Last Year')
plt.xlabel('Late Payments Last Year')
plt.ylabel('Frequency')
plt.show()
```
